```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11m.pt")

# Train the model
results = model.train(data="brain-tumor.yaml", epochs=10, imgsz=640, device=0, batch=16,
    patience=15 # Early stopping if no improvement
                      )
```

```python
results = model.val(
    data="brain-tumor.yaml",
    imgsz=640,
    batch=32,
    conf=0.001,
    iou=0.7,
    device=0
)
```

```python
results = model.predict(
    source="/content/datasets/brain-tumor/valid/images/val_1 (1).jpg",
    conf=0.1,
    imgsz=1280,
    device=0,
    save=True  # Saves to runs/detect/predict/
)
```

```python
# Install the required package (Supervision)
!pip install supervision

# Import necessary libraries
import supervision as sv
import cv2
import random
import numpy as np

# List all image paths from your dataset
image_paths = sv.list_files_with_extensions(
    directory="/content/datasets/brain-tumor/valid/images",
    extensions=['png', 'jpg', 'jpeg']
)

# Choose a random image from the dataset
image_path = random.choice(image_paths)
image = cv2.imread(str(image_path))

# Make predictions using the model
results = model.predict(image, conf=0.10)[0]

# Convert the results into a Detections object
detections = sv.Detections.from_ultralytics(results)

# Check if there are any detections and keep only the highest confidence one
if len(detections) > 0:
    # Find the index of the detection with the highest confidence score
    highest_conf_index = detections.confidence.argmax()

    # Get the class id, confidence, and bounding box for the top detection
    top_confidence = detections.confidence[highest_conf_index]
    top_class_id = detections.class_id[highest_conf_index]
    top_bbox = detections.xyxy[highest_conf_index]

    # Convert confidence to percentage
    confidence_percentage = f"{top_confidence * 100:.2f}%"

    # Keep only the highest confidence detection
    detections = sv.Detections(
        xyxy=np.array([top_bbox]),  # 2D array for bounding box
        confidence=np.array([top_confidence]),  # 1D array for confidence
        class_id=np.array([top_class_id])  # 1D array for class_id
    )

    # Annotate the image with the top detection
    box_annotator = sv.BoundingBoxAnnotator()
    label_annotator = sv.LabelAnnotator(text_color=sv.Color.BLACK)

    # Create a label with class name and confidence percentage
    class_name = model.names[top_class_id]  # Get class name from model
    label = f"{class_name} {confidence_percentage}"

    # Annotate with both bounding box and label
    annotated_image = box_annotator.annotate(scene=image.copy(), detections=detections)
    annotated_image = label_annotator.annotate(scene=annotated_image, detections=detections, labels=[label])

    # Plot the annotated image
    sv.plot_image(annotated_image)
else:
    print("No detections found.")
```
